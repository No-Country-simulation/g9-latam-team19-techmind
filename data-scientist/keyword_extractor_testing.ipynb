# 🔎 Keyword Extractor - Testing Notebook

This notebook validates the keyword extraction pipeline.

The pipeline:
1. Receives a document.
2. Extracts and cleans text.
3. Detects keywords using the catalog.
4. Generates a JSON output.


```python
from keyword_functions import (
    extract_text,
    clean_text,
    load_keyword_catalog,
    extract_keywords,
    save_keywords_json
)

from extractor import process_data
```

```python
#se pueden probar distintos ejemplos modificando data
data = {
    "title": "Angular: Controla la navegación",
    "text": "Aprende componentes reutilizables con spring boot y python"
}
```

```python
#usando las funciones sueltas

raw_text = extract_text(data)

cleaned_text = clean_text(raw_text)

catalog = load_keyword_catalog()

keywords = extract_keywords(cleaned_text, catalog)

save_keywords_json(keywords)

print(keywords)
```
